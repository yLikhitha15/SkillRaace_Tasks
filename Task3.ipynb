```python
  #Dependencies
  import numpy as np
  import pandas as pd
  from sklearn.preprocessing import StandardScaler
  from sklearn.model_selection import train_test_split
  from sklearn import svm
  from sklearn.metrics import accuracy_score
```

```python
#Data Collection and Analysis
diabetesdataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetesdataset.head()
```

```python
diabetesdataset.shape
```

```python
#Statistical measure of the data (mean, standard deviation)
diabetesdataset.describe()
```

```python
diabetesdataset['Outcome'].value_counts()
```

```python
#0-Non Diabetic, 1-Diabetic
diabetesdataset.groupby('Outcome').mean()
```

```python
#Separating data and labels
X=diabetesdataset.drop(columns='Outcome', axis=1)
Y=diabetesdataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Standardize the data i.e, Data Standardization
scaler=StandardScaler()
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X=standardized_data
Y=diabetesdataset['Outcome']
print(X)
print(Y)
```

```python
#Train Test Split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Train the model
classifier=svm.SVC(kernel='linear')
```

```python
#Training the support vector machine classifier
classifier.fit(X_train,Y_train)
```

```python
#Model Evaluation
#Accuracy Score
#Accuracy score on the training data
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy Score of the Training Data = ',training_data_accuracy)
```

```python
#Accuracy Score on the testing data
X_test_prediction=classifier.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy Score of the Testing Data = ',testing_data_accuracy)
```

```python
#Making the Predictive System
input_data=(4,110,92,0,0,37.6,0.191,30)
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

#Standardize the input data
scaler = StandardScaler(copy=False)
scaler.fit(X)
std_data=scaler.transform(input_data_reshaped)
print('Standard Data: ',std_data)

prediction=classifier.predict(std_data)
if (prediction[0]==0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
